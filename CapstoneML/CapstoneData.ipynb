```python
data["EXPENSE_PER_STUDENT"] = data["INSTRUCTION_EXPENDITURE"]/data["ENROLL"]
```

```python
data['EXPENSE_PER_STUDENT'].replace(np.inf,np.nan, inplace = True)
```

```python
mean_value=data['EXPENSE_PER_STUDENT'].mean()
data['EXPENSE_PER_STUDENT'].fillna(value = mean_value, inplace=True)
```
